```python
input = open("./../../data/day_4/input.txt", "r").readlines()
```

```python
#process inputs
cards = {}
for line in input:
    name, content = line.split(":")
    name = re.search("[A-z]* ([0-9]+)", name).group(1)
    winning, numbers = content.split("|")
    winning = winning.strip().split()
    numbers = numbers.strip().split()
    cards[int(name)] = {
        "winning": [int(x) for x in winning],
        "numbers": [int(x) for x in numbers]
    }
```

```python
# part 1
points = []
for k,card in cards.items():
    winning = card["winning"]
    numbers = card["numbers"]
    correct = 0
    for num in numbers:
        if num in winning:
            correct += 1
    points.append(2**(correct-1) if correct > 0 else 0)

total_points = sum(points)
print(f"Part 1: {total_points}")
```

```python
# part 2
instances = {
    k: 1 for k in cards
}
num_cards = len(cards)
for k,card in cards.items():
    winning = card["winning"]
    numbers = card["numbers"]
    correct = 0
    for num in numbers:
        if num in winning:
            correct += 1
    if correct == 0:
        continue
    for copy in range(k+1, k+correct+1):
        if copy in instances:
            instances[copy] += instances[k]
            print(f"Adding {instances[k]} copies of card {copy} because of card {k} which has {correct} correct")

total_cards = sum(instances.values())
print(f"Part 2: {total_cards}")
```
